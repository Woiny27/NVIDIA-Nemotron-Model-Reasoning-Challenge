# Secure Git push from Colab

This Colab notebook contains ready-to-run cells to securely push commits to your repository without writing your Personal Access Token (PAT) to disk. It implements the recommended `http.extraheader` flow and an optional `gh` (GitHub CLI) flow.

Important: update the repo remote URL below if your repository NWO differs from Woiny27/NVIDIA-Nemotron-Model-Reasoning-Challenge. Do NOT paste tokens directly into cells.

```python
# Configure git identity (one-time per Colab runtime)
!git config --global user.email "you@example.com"
!git config --global user.name "Your Name"
```

```python
# Verify or set the correct remote (edit the URL if needed)
!git remote -v
# If remote is missing/wrong, uncomment and run the next line (replace the URL if needed)
# !git remote set-url origin https://github.com/Woiny27/NVIDIA-Nemotron-Model-Reasoning-Challenge.git
```

```python
# Securely read PAT into memory (will NOT be saved to disk)
from getpass import getpass
import os
os.environ['GITHUB_TOKEN'] = getpass('GitHub token (will not be saved to disk): ')
```

```python
# Commit and push using the token via an HTTP header (safer than embedding in URL)
!git add .
!git commit -m "Updated reasoning workflow" || echo "No changes to commit"
!git -c http.extraheader="AUTHORIZATION: bearer $GITHUB_TOKEN" push origin main
```

Optional: GitHub CLI flow (alternative)


```python
# Install GitHub CLI in Colab (may take ~30s)
!curl -fsSL https://cli.github.com/packages/githubcli-archive-keyring.gpg | sudo dd of=/usr/share/keyrings/githubcli-archive-keyring.gpg
!sudo chmod go+r /usr/share/keyrings/githubcli-archive-keyring.gpg
!echo "deb [arch=$(dpkg --print-architecture) signed-by=/usr/share/keyrings/githubcli-archive-keyring.gpg] https://cli.github.com/packages stable main" | sudo tee /etc/apt/sources.list.d/github-cli.list > /dev/null
!sudo apt update
!sudo apt install gh -y
```

```python
# Authenticate gh using a token supplied via getpass (will NOT be saved to disk)
from getpass import getpass
import subprocess

token = getpass('GitHub token (will not be saved to disk): ')
proc = subprocess.run(['gh','auth','login','--with-token'], input=token.encode())
print('gh auth exit code:', proc.returncode)
# After successful auth, run the usual git add/commit/push commands in bash cells
```
